## Train, test split equally

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(Hdata, Hdata['CHAS']):
    train_set = Hdata.loc[train_index]
    test_set = Hdata.loc[test_index]
```

## facing the null value

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy="median")
imputer.fit(Hdata)
```

```python
x_train_main = train_set.drop('MEDV', axis=1)
y_train = train_set['MEDV']
```

```python
x_train_main
```

## creating a pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    ('std_scaler', StandardScaler())
])
```

```python
x_train = my_pipeline.fit_transform(x_train_main)
```

```python
x_train.shape
```

## Start prediction

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

# my_model = LinearRegression()
# my_model = DecisionTreeRegressor()
my_model = RandomForestRegressor()

my_model.fit(x_train, y_train)
```

* take some data and label for predict from train data

```python
some_data = x_train_main.iloc[:5]
some_label = y_train.iloc[:5]
```

```python
pre_some_data = my_pipeline.fit_transform(some_data)
```

```python
pred_some_data = my_model.predict(pre_some_data)
```

```python
list(some_label)
```

## Find Rootover Mean Squared Error

```python
from sklearn.metrics import mean_squared_error
y_train_pred = my_model.predict(x_train)
mse = mean_squared_error(y_train, y_train_pred)
```

```python
rmse = np.sqrt(mse)
```

```python
rmse
```

## Using better evaluation technique / finding mean error - Cross Validation

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(my_model, x_train, y_train, scoring='neg_mean_squared_error', cv=10)
scores_rmse = np.sqrt( - scores ) # - for because score value got in negative number
```

```python
scores_rmse
```

```python
def print_crossVal(dt):
    print("Cross_Val_ Rootover Mean Squared Error :", dt)
    print("Mean :", dt.mean())
    print("Standard Value :", dt.std())
    
```

```python
print_crossVal(scores_rmse)
```
